```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Will import and start analyzing the dataset
logs=pd.read_csv('log.csv')
```

```python
logs.head(10)
```

```python
#Will check if there is any null or NaN values in file
logs.isnull()
```

```python
logs.isnull().sum()
```

```python
#Perfect, let's start with plotting
```

```python
plt.subplot(1, 3, 1)
plt.plot(logs['GR'].values, logs['Depth'].values, 'g', label='Gamma Ray')
plt.title('Gamma Ray')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid(True)
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)


plt.subplot(1, 3, 2)
plt.plot(logs['RLL3'].values, logs['Depth'].values,'tomato', label='Resistivity')
plt.title('Resistivity')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid()
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)


plt.subplot(1, 3, 3)
plt.plot(logs['CNLS'].values, logs['Depth'].values,'dimgrey', label='Neutron Porosity')
plt.plot(logs['DPOR'].values, logs['Depth'].values,'gold', label='Density')
plt.title('Neutron Porosity-Density')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid()
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)
```

```python
plt.subplot(1, 3, 1)
plt.plot(logs['SP'].values, logs['Depth'].values, 'goldenrod', label='Spontaneous Potential')
plt.title('Spontaneous Potential')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid(True)
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)


plt.subplot(1, 3, 2)
plt.plot(logs['DCAL'].values, logs['Depth'].values,'c', label='Calliper Log')
plt.title('Calliper Log')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid()
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)


plt.subplot(1, 3, 3)
plt.plot(logs['RHOB'].values, logs['Depth'].values,'peru', label='RHOB')
plt.title('Density')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid()
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)
```

```python
plt.subplot(1, 3, 1)
plt.plot(logs['RILD'].values, logs['Depth'].values, 'brown', label='Induction')
plt.title('Induction')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid(True)
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)


plt.subplot(1, 3, 2)
plt.plot(logs['MCAL'].values, logs['Depth'].values,'gray', label='Micro Caliper')
plt.plot(logs['MN'].values, logs['Depth'].values,'lime', label='Micro Normal')
#plt.plot(logs['MI'].values, logs['Depth'].values,'whitesmoke', label='Micro Inverse')
plt.title('Micro-Caliper and Normal')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid()
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)


plt.subplot(1, 3, 3)
plt.plot(logs['RHOC'].values, logs['Depth'].values,'peru', label='RHOC')
plt.title('Corrected Bulk density')
plt.yticks(np.arange(0, max(logs['Depth'].values), 500.0))
plt.grid()
plt.legend(fontsize=13.5, loc=1)
plt.gca().invert_yaxis()
plt.gcf().set_size_inches(20, 20)
```
